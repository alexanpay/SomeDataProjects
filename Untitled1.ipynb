# Data Manipulation Project

### Importing the first file
First file is the energy data from the file `Energy Indicators.xls`, which is a list of indicators of [energy supply and renewable electricity production](Energy%20Indicators.xls) from the [United Nations](http://unstats.un.org/unsd/environment/excel_file_tables/2013/Energy%20Indicators.xls) for the year 2013

```python
import pandas as pd
import numpy as np

#First data "Energy Indicators.xls" from UN.
energy = pd.read_excel(r"C:\Users\alexa\Desktop\Energy Indicators.xls",skip_footer=38,skip_header=1,skiprows=17)

#Rename columns to more appropriate ones
energy = energy.rename(columns = {"Unnamed: 1" : "Country", "Petajoules" : "Energy Supply", "Gigajoules" : "Energy Supply per capita", '%' : '% Renewable'})

#Drop unnecessary columns (contains fully NA) 
energy.drop(["Unnamed: 0", "Unnamed: 2"], axis = 1, inplace = True)

#Drop rows with NA values
energy.dropna()

#Rename some values in column "Country"
energy['Country'] = energy['Country'].str.replace('Republic of Korea','South Korea')
energy['Country'] = energy['Country'].str.replace('United States of America', 'United States')
energy['Country'] = energy['Country'].str.replace('United Kingdom of Great Britain and Northern Ireland', 'United Kingdom')
energy['Country'] = energy['Country'].str.replace('China, Hong Kong Special Administrative Region', 'Hong Kong')

#Replace '...' to 'NaN' in values
for col in energy:
    energy[col] = energy[col].replace('...',np.nan)

#Converting Energy to gigajoules
energy['Energy Supply'] *= 1000000

energy.head()
```

### Importing the second file

Next file is the GDP data from the file `world_bank.csv`, which is a csv containing countries' GDP from 1960 to 2019 from [World Bank](http://data.worldbank.org/indicator/NY.GDP.MKTP.CD).


```python
#Second data "world_bank.csv"
GDP = pd.read_csv(r"C:\world_bank.csv", skiprows = 4)

#Rename some countries so it will fit with the first data
GDP["Country Name"] = GDP["Country Name"].str.replace('Korea, Rep.', 'South Korea')
GDP["Country Name"] = GDP["Country Name"].str.replace('Iran, Islamic Rep.', 'Iran')
GDP["Country Name"] = GDP["Country Name"].str.replace("Hong Kong SAR, China", "Hong Kong")

#Rename "Country Name" column to a "Country"
GDP = GDP.rename(columns = {"Country Name" : "Country"})

#There are so many years. Let's drop some years columns and save only period from 2006 - 2015
GDP.drop(GDP.iloc[:, 4:50], inplace = True, axis = 1) 
GDP.drop(GDP.iloc[:, 14:19], inplace = True, axis = 1)

GDP.head()
```

### Importing the third file
Finally, third data is the [Sciamgo Journal and Country Rank data for Energy Engineering and Power Technology](http://www.scimagojr.com/countryrank.php?category=2102) from the file `scimagojr.xlsx`, which ranks countries based on their journal contributions in the aforementioned area.

```python
#Third data "scimagojr country rank 1996-2019.xlsx"
ScimEn = pd.read_excel(r"C:\scimagojr.xlsx")

#Merging all dataframes
new_df = pd.merge(energy, GDP, how='outer', left_on = 'Country', right_on="Country")
new_df2 = pd.merge(new_df, ScimEn, how='outer', left_on = "Country", right_on = "Country")

#Drop unnecessary columns
new_df2.drop(["Country Code", "Indicator Code", "Indicator Name", "Region"], axis = 1, inplace = True)

#Drop NA in 'Rank' column
new_df2 = new_df2[new_df2['Rank'].notna()]

#Modify value in "Rank" from float to int since we will set them as index
new_df2[["Rank"]] = new_df2[["Rank"]].astype(int)

#Setting column 'Rank' as index
new_df2 = new_df2.set_index("Country")

#Sorting index in ascending order
new_df2 = new_df2.sort_values("Rank")

#Top 15 countries by ranking
Top15 = new_df2[:15]
Top15
```

### The average GDP for the last 10 years for each country

###  Dateframe that displays the sample size (the number of countries in each continent bin), and the sum, mean, and std deviation for the estimated population of each country

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America',
                  'Netherlands':'Europe'}

Top15['Continent'] = Top15.index.to_series().map(ContinentDict)
Top15['Population Estimate'] = (Top15['Energy Supply']/Top15['Energy Supply per capita']).astype(float)
dff = Top15.set_index('Continent').groupby(level = 0)['Population Estimate'].agg(**{'size':np.size, 'sum':np.sum, 'mean':np.mean, 'std':np.std})
    
dff
```
